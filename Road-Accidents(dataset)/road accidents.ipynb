```python
import pandas as pd


cities = pd.read_csv("Accident_Cities.csv")
states = pd.read_csv("Accident_States.csv",skiprows=1)
states.head()
```

```python
cities=cities.fillna(0)
cities.head()
```

```python
cities["total"] = cities["Persons Injured - 2011"]+cities["Persons Injured - 2012"]+cities["Injured - 2013"]+cities["Injured - 2014"]+cities["Injured - 2015"]
```

```python
cities.head()
```

```python
cities = cities.sort_values("total",ascending=False)

cities["Name of City"].head(3)
```
