```python
import panda
df1=pd.read_csv("milk_production.csv")
df1=df1.fillna(0)
```

```python
# df1.loc[:,[["Cow Milk-2013-14","Boffalo Milk-2013-14","Goat Milk-2013-14"]]]
df1["max13"]=df1.iloc[:,[4,9,14]].sum(axis=1)
df1.sort_values("max13",ascending=False).head(1)["State/ UT Name"]
```

```python
# df1.loc[:,"Goat Milk-2013-14"]
df1.columns
```

```python
df1.head()
```

```python
df1["max10"]=df1.iloc[:,[2,7,12]].sum(axis=1)
df1["max11"]=df1.iloc[:,[3,8,13]].sum(axis=1)
df1["max14"]=df1.iloc[:,[5,10,15]].sum(axis=1)
df1["max15"]=df1.iloc[:,[6,11,16]].sum(axis=1)

print(df1.sort_values(["max10"],ascending=False).head()["State/ UT Name"])
print(df1.sort_values(["max11"],ascending=False).head()["State/ UT Name"])
df1.sort_values(["max15"],ascending=False).head()["State/ UT Name"]
df1.sort_values(["max14"],ascending=False).head()["State/ UT Name"]
```

```python
df1["total"]=df1.iloc[:,17:22].sum(axis=1)

df1["cowtot"]=df1.iloc[:,2:7].mean(axis=1)
df1["boftot"]=df1.iloc[:,7:12].mean(axis=1)
df1["totgoat"]=df1.iloc[:,12:17].mean(axis=1)

def avg(state):
    print(df1[df1["State/ UT Name"]==state][["cowtot","boftot","totgoat"]])
    
avg("Andhra Pradesh")
```

```python
df1.columns
```

```python
def graph(state):
    y=df1[df1["State/ UT Name"]==state][["max10","max11","max13","max14","max15"]].values.ravel()
    x=np.array(["2010","2011","2013","2014","2015"])
    plt.plot(x,y)
    plt.show()
graph("Arunachal Pradesh")
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

def plott(state):
    x1=df1[df1["State/ UT Name"]==state][["Cow Milk-2011-12","Boffalo Milk-2011-12","Goat Milk-2011-12"]].values.ravel()
    x2=df1[df1["State/ UT Name"]==state][["Cow Milk-2013-14","Boffalo Milk-2013-14","Goat Milk-2013-14"]].values.ravel()
    x3=df1[df1["State/ UT Name"]==state][["Cow Milk-2014-15","Boffalo Milk-2014-15","Goat Milk-2014-15"]].values.ravel()
    x4=df1[df1["State/ UT Name"]==state][["Cow Milk-2015-16","Boffalo Milk-2015-16","Goat Milk-2015-16"]].values.ravel()
    ax1.pie(x1)                                     
    ax2.pie(x2,autopct="%1.0f%%")       
    ax3.pie(x3,autopct="%1.0f%%")
    ax4.pie(x4,autopct="%1.0f%%")
    plt.show()
                                         
plott("Gujarat")
```
