```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

milkDF = pd.read_csv("milk_production.csv")
milkDF=milkDF.fillna(0)
milkDF.head()
```

```python
#1 Which state has produced max milk in year 2013-14?
milkDF["maxmilk2013_14"] = milkDF["Cow Milk-2013-14"] + milkDF["Boffalo Milk-2013-14"] + milkDF["Goat Milk-2013-14"]

milkDF[milkDF["maxmilk2013_14"]==milkDF["maxmilk2013_14"].max()]["State/ UT Name"]
```

```python
#2.List top 5 milk producing states in each year from 2010 to 2015

milkDF["y2011"]=milkDF["Cow Milk-2010-11"] + milkDF["Boffalo Milk-2010-11"] + milkDF["Goat Milk-2010-11"]
milkDF["y2012"]=milkDF["Cow Milk-2011-12"] + milkDF["Boffalo Milk-2011-12"] + milkDF["Goat Milk-2011-12"]
milkDF["y2014"]=milkDF["Cow Milk-2013-14"] + milkDF["Boffalo Milk-2013-14"] + milkDF["Goat Milk-2013-14"]
milkDF["y2015"]=milkDF["Cow Milk-2014-15"] + milkDF["Boffalo Milk-2014-15"] + milkDF["Goat Milk-2014-15"]
milkDF["y2016"]=milkDF["Cow Milk-2015-16"] + milkDF["Boffalo Milk-2015-16"] + milkDF["Goat Milk-2015-16"]

print("2011 top 5states")
print(milkDF.sort_values("y2011",ascending=False)["State/ UT Name"].head())
print("2012 top 5states")
print(milkDF.sort_values("y2012",ascending=False)["State/ UT Name"].head())
print("2014 top 5states")
print(milkDF.sort_values("y2014",ascending=False)["State/ UT Name"].head())
print("2015 top 5states")
print(milkDF.sort_values("y2015",ascending=False)["State/ UT Name"].head())
print("2016 top 5states")
print(milkDF.sort_values("y2016",ascending=False)["State/ UT Name"].head())
```

```python
#3. Calculate average milk production by cow, buffalo and goat in each year for selected state.

def average(state,year):
    #print(milkDF["y2011"].mean())
    if year == "2011":
        print(milkDF[milkDF["State/ UT Name"]==state]["y2011"]/3)
    if year == "2012":
        print(milkDF[milkDF["State/ UT Name"]==state]["y2012"]/3)
    if year == "2014":
        print(milkDF[milkDF["State/ UT Name"]==state]["y2014"]/3)    
    
average("Rajasthan","2011")
```

```python
#4. List 5 states (if present) whose total milk production is increased for last three years.

milkDF[milkDF["y2016"].ge(milkDF["y2015"]) & milkDF["y2015"].ge(milkDF["y2014"])].head()
```

```python
#5. Draw line graph showing total milk production for last 5 years of selected state
x = [1,2,3,4,5]


y = [milkDF.get_value(0,"y2011"),milkDF.get_value(0,"y2012"),milkDF.get_value(0,"y2014"),milkDF.get_value(0,"y2015"),milkDF.get_value(0,"y2016")]
plt.plot(x,y,label="line graph for milk production of AP",marker="o")
#plt.xticks(["2011","2012","2014","2015","2016"])
plt.xlabel("last 5years")
plt.ylabel("milk production")
plt.legend()

plt.show()
```

```python
#6. Draw subplots of pie charts showing % milk productions by cow, buffalo and goat for last 4 years of selected state.
#milkDF["cow"]= milkDF[""]+milkDF["Cow Milk-2013-14"]+milkDF["Cow Milk-2014-15"]+milkDF["Cow Milk-2015-16"]
cow = [milkDF.get_value(0,"Cow Milk-2011-12"),milkDF.get_value(0,"Cow Milk-2013-14"),milkDF.get_value(0,"Cow Milk-2014-15"),milkDF.get_value(0,"Cow Milk-2015-16")]
boffalo = [milkDF.get_value(0,"Boffalo Milk-2011-12"),milkDF.get_value(0,"Boffalo Milk-2013-14"),milkDF.get_value(0,"Boffalo Milk-2014-15"),milkDF.get_value(0,"Boffalo Milk-2015-16")]
goat =[milkDF.get_value(0,"Goat Milk-2011-12"),milkDF.get_value(0,"Goat Milk-2013-14"),milkDF.get_value(0,"Goat Milk-2014-15"),milkDF.get_value(0,"Goat Milk-2015-16")]

fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(wspace = 0.5, hspace = 0.3)

axes[0,0].pie(cow,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[0,0].set_title("cow production")

axes[0,1].pie(boffalo,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[0,1].set_title("Boffalo production")

axes[1,0].pie(goat,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[1,0].set_title("Goat production")
plt.show()
```
